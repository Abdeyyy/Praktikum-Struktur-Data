```python
#COIN CHANGE

def coin_change_greedy(amount, coins):
  coins.sort(reverse=True) #mengurutkan dari terbesar
  result = [] #list untuk menyimpan coin yang digunakan
  for coin in coins: #iterasi tiap koin
    while amount >= coin: #selama koin lebih besar dari amount
      amount -= coin #amount dikurangi koin saat ini
      result.append(coin) #koin ditambahkan ke dalam result
  return result #mengembalikan list result

amount = 57
coins = [25, 10, 5, 1]
change = coin_change_greedy(amount, coins)
print('Koin yang digunakan: ', change)
```

```python
#KNAPSACK

def fractional_knapsack(items, capacity):
  #items = list of tuples (value, weight)
  items = sorted(items, key=lambda x: x[0]/x[1], reverse=True) #urutkan dari rasio terbesar - terkecil
  total_value = 0.0 #variabel untuk menyimpan total nilai yang ditampung
  for value, weight in items: #iterasi items
    if capacity >= weight: #jika kapasitas masih ada
      capacity -= weight #kapasitas dikurangi berat dari nilai
      total_value += value #menambahkan nilai ke  variabel total_value
    else: #jika kapasitas tidak sebesar berat dari nilai saat ini
      total_value += value * (capacity/weight)  #tambahkan sebagian nilai sesuai sisa kapasitas
      break
  return total_value #kembalikan nilai variabel total_value

items =[(60, 10), (100, 20), (120, 30)] #(nilai, berat)
capacity = 50
max_value = fractional_knapsack(items, capacity)
print('Nilai maksimum yang dapat dibawa:', max_value)
```

```python
#PRIM

import heapq

def prim_mst(graph, start):
  visited = set() #simpan simpul yang dikunjungi
  min_heap = [(0, start)] #heap min (bobot, simpul)
  total_weight = 0 #total bobot MST

  while min_heap:
    weight, node = heapq.heappop(min_heap) #ambil bobot terkecil
    if node in visited:
      continue
    visited.add(node) #tandai simpul yang dikunjungi
    total_weight += weight #tambahkan bobot ke total_weight
    for neighbor, edge_weight in graph[node]: #cek tetangga simpul
      if neighbor not in visited:
        heapq.heappush(min_heap, (edge_weight, neighbor)) #tambahkan ke heap

  return total_weight #mengembalikan total bobot MST

graph = {
    'A' : [('B', 2), ('C', 3)],
    'B' : [('A', 2), ('C', 1), ('D', 1)],
    'C' : [('A', 3), ('B', 1), ('D', 4)],
    'D' : [('B', 1), ('C', 4)]
}

print('Total bobot MST (Prim):', prim_mst(graph, 'A'))
```

```python
#KRUSKAL

def kruskal_mst(edges, n_nodes):
  parent = {i : i for i in range(n_nodes)} #inisialisasi dict parent

  def find(x):
    while parent[x] != x: #selama x bukan parent
      x = parent[x] #naik ke atas hingga menemukan root
    return x #mengembalikan nilai root

  def union(x, y):
    root_x = find(x) #cari root x
    root_y = find(y) #cari root y
    if root_x != root_y: #jika root berbeda
      parent[root_y] = root_x #gabungkan keduanya
      return True #berhasil digabung
    return False #sudah satu kesatuan

  edges.sort(key=lambda x: x[2]) #urutkan sisi berdasarkan bobot terkecil
  total_weight = 0 #bobot awal MST
  for u, v, weight in edges: #iterasi semua sisi yang diurutkan
    if union(u, v): #jika sisi tidak membentuk siklus
      total_weight += weight #tambah bobot ke total_weight

  return total_weight #mengembalikan nilai total_weight

#edges: (node1, node2, weight), nodes are represented by integers
edges = [(0, 1, 2), (0, 2, 3), (1, 2, 1), (1, 3, 1), (2, 3, 4)]
print('Total bobot MST (Kruskal):', kruskal_mst(edges, 4))
```
