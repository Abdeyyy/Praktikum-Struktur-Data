```python
#LINEAR SEARCH

def linear(arr, target):
  for i in range(len(arr)):
    if arr[i]== target:
      return i
  return -1

arr = [12,16,37,97,16,90,17,42,67]
target = int(input('Masukkan 2 digit terakhir NPM anda : '))
result = linear(arr, target)

if result !=1:
  print(f'Elemen ditemukan pada indeks {result}')
else:
  print('Elemen tidak ditemukan')
```

```python
#BINARY SEARCH

def binary(arr, target):
  low = 0
  high = len(arr)-1

  while low <= high:
    mid = (low+high)//2
    if arr[mid] == target:
      return mid
    elif arr[mid] < target:
      low = mid+1
    else:
      high = mid-1
  return -1

arr = [1001,1291,7462,8472,9271,9373,9384]
target = int(input('Masukkan tanggal lahir anda : '))
result = binary(arr,target)

if result !=1:
  print(f'Elemen ditemukan pada indeks {result}')
else:
  print('Elemen tidak ditemukan')
```

```python
#JUMP SEARCH

import math

def jump(arr, target):
  n = len(arr)
  step = int(math.sqrt(n))
  prev = 0

  while arr[min(step,n)-1] < target:
    prev =  step
    step+= int(math.sqrt(n))
    if prev >= n:
      return -1

  for i in range(prev, min(step,n)):
    if arr[i] == target:
      return i
  return -1

arr = ['Abdi','Bayu','Cia','Diana']
target = input('Masukkan nama anda : ')
result = jump(arr, target)

if result !=-1:
  print(f'Elemen ditemukan pada indeks {result}')
else:
  print('Elemen tidak ditemukan')
```

```python
import random

posisi = -1
def linear_search(arr, target):
  global posisi
  for i in range(len(arr)-1):
    if arr[i] == target:
      posisi = i
      return True
  return False

def binary_search(arr, target):
  low = 0
  high = len(arr)-1
  global posisi
  while low <= high:
    mid = (low+high)//2
    if arr[mid] == target:
      posisi = mid
      return True
    elif arr[mid] < target:
      low = mid+1
    else:
      high = mid-1

def jump_search(arr, target):
  n = len(arr)
  step = int(math.sqrt(n))
  prev = 0
  global posisi

  while arr[min(step,n)-1] < target:
    prev = step
    step += int(math.sqrt(n))
    if prev >= n:
      return False
  for i in range(prev, min(step,n)):
    if arr[i]==target:
      posisi = i
      return True
  return False

n = random.sample(range(1,1001), 100)
arr = sorted(n)
print(arr)
print(20*'='+'Searching List'+20*'=')
while True:
  print('Masukkan pilihan : ')
  pilihan = int(input('1. Linear Search\n2. Binary Search\n3. Jump Search\n4. Keluar\n'))
  if pilihan == 1:
    target = int(input('Masukkan angka yang dicari : '))
    linear_search(arr, target)
    if linear_search(arr, target):
      print(f'Elemen ditemukan pada indeks {posisi}\n')
    else:
      print('Elemen tidak ditemukan!\n')
  elif pilihan == 2:
    target = int(input('Masukkan angka yang dicari : '))
    binary_search(arr, target)
    if binary_search(arr, target):
      print(f'Elemen ditemukan pada indeks {posisi}\n')
    else:
      print('Elemen tidak ditemukan!\n')
  elif pilihan == 3:
    target = int(input('Masukkan angka yang dicari : '))
    jump_search(arr, target)
    if jump_search(arr, target):
      print(f'Elemen ditemukan pada indeks {posisi}\n')
    else:
      print('Elemen tidak ditemukan!\n')
  elif pilihan == 4:
    break
  else:
    print('Perintah tidak ada')
```

```python
import random
import math
import tabulate

def linear_search(arr, target):
  i = 0
  while i < len(arr):
    i +=1
    if arr[i] == target:
      return i+1
  return i

def binary_search(arr, target):
  low = 0
  high = len(arr)-1
  i = 0
  while low <= high:
    i += 1
    mid = (low+high)//2
    if arr[mid] == target:
      posisi = mid
      return i
    elif arr[mid] < target:
      low = mid+1
    else:
      high = mid-1

def jump_search(arr, target):
  n = len(arr)
  step = int(math.sqrt(n))
  prev = 0
  i = 0
  while arr[min(step,n)-1] < target:
    i+=1
    prev = step
    step += int(math.sqrt(n))
    if prev >= n:
      return i
  for j in range(prev, min(step,n)):
    i+=1
    if arr[j]==target:
      return i
  return i

arr = random.sample(range(1, 1001), 100)
arr = sorted(arr)
target = random.choice(arr)
print('Target :',target)
print('Indeks target :',arr.index(target))
linear = linear_search(arr, target)
binary = binary_search(arr, target)
jump = jump_search(arr, target)

data =[['Linear Search', linear],
       ['Binary Search', binary],
       ['Jump Search', jump]

]

print(tabulate.tabulate(data, headers=['Algoritma Pencarian', 'Jumlah perbandingan'], tablefmt='fancy_grid'))
```
