```python
#BUBBLE SORT

def bubble_sort(arr):
  n = len(arr)
  for i in range(n-1):
    for j in range(n-i-1):
      if arr[j]>arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
  return arr

data = [64, 34, 25, 12, 22,11,90]
sorted = bubble_sort(data.copy())
print('Sebelum urut :',data)
print('Setelah urut :',sorted)
```

```python
#SELECTION SORT

def selection_sort(arr):
  n = len(arr)
  for i in range(n-1):
    min_index = i
    for j in range(i +1, n):
      if arr[j] < arr[min_index]:
        min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]
  return arr

data = [64,25,12,22,11]
sorted = selection_sort(data.copy())
print('Sebelum urut :',data)
print('Sesudah urut :',sorted)
```

```python
#INSERTION SORT

def insertion_sort(arr):
  n = len(arr)
  for i in range(1, n):
    key = arr[i]
    j = i -1
    while j >= 0 and arr[j]>key:
      arr[j+1] = arr[j]
      j-=1
    arr[j+1]=key
  return arr

data = [64,34,25,12,22,11,90]
sorted = insertion_sort(data.copy())
print('Sebelum urut :',data)
print('Setelah urut :',sorted)
```

```python
#MERGE SORT

def merge_sort(arr):
  if len(arr)>1:
    mid = len(arr)//2
    left_half = arr[:mid]
    right_half = arr[mid:]

    merge_sort(left_half)
    merge_sort(right_half)

    i=j=k=0

    while i < len(left_half) and j < len(right_half):
      if left_half[i] < right_half[j]:
        arr[k] = left_half[i]
        i +=1
      else:
        arr[k] = right_half[j]
        j +=1
      k+=1

    while i < len(left_half):
      arr[k] = left_half[i]
      i +=1
      k +=1

    while j < len(right_half):
      arr[k] = right_half[j]
      j += 1
      k += 1

  return arr

data = [64, 25,12, 22,11]
sorted = merge_sort(data.copy())
print('Sebelum urut :', data)
print('Setelah urut :', sorted)
```

```python
#QUICK SORT

def quick_sort(arr):
  if len(arr) <= 1:
    return arr

  pivot = arr[len(arr)//2]
  left = [x for x in arr if x < pivot]
  middle = [x for x in arr if x == pivot]
  right = [x for x in arr if x > pivot]

  return quick_sort(left) + middle + quick_sort(right)

data = [64,25,12,22,11]
sorted = quick_sort(data.copy())
print('Sebelum urut :', data)
print('Setelah urut :', sorted)
```

```python
class Data:
  def __init__(self):
    self.nama = []
    self.umur = []
    self.total = []

  def tambah_nama(self, nama):
    self.nama.append(nama)
    return self.nama

  def tambah_umur(self, umur):
    self.umur.append(umur)
    return self.umur

  def tambah_total(self, total):
    self.total.append(total)
    return self.total

  def urut_nama(self):
    return self.quick_sort(self.nama)

  def urut_umur(self):
    return self.quick_sort(self.umur)

  def urut_total(self):
    return self.quick_sort(self.total)

  def quick_sort(self, arr):
    if len(arr) <= 1:
      return arr
    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    mid = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return self.quick_sort(left) + mid + self.quick_sort(right)

data = Data()
data.tambah_nama('Bagas')
data.tambah_nama('Sabil')
data.tambah_nama('Abdi')
data.tambah_nama('Hammam')
data.tambah_nama('Nabila')
data.tambah_nama('Yasa')
data.tambah_nama('Rayfal')
print('Nama sebelum terurut : ', data.tambah_nama('Ilyasa'))
print('Nama sesudah terurut :', data.urut_nama())
print('')
data.tambah_umur(20)
data.tambah_umur(10)
data.tambah_umur(30)
data.tambah_umur(24)
data.tambah_umur(27)
print('Umur sebelum terurut : ',data.tambah_umur(6))
print('Umur setelah terurut : ',data.urut_umur())
print('')
data.tambah_total(14000)
data.tambah_total(50000)
data.tambah_total(30000)
data.tambah_total(10000)
data.tambah_total(70000)
data.tambah_total(13000)
print('Total harga sebelum terurut : ', data.tambah_total(15000))
print('Total harga setelah terurut : ', data.urut_total())
```
