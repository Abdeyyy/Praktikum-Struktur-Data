```python
class TextEditor:
  def __init__(self):
    self.history = []

  def write(self, text):
    self.history.append(text)

  def undo(self):
    if self.history:
      return f"Undo menghapus '{self.history.pop()}'"
    return 'Tidak ada yang bisa di undo'


editor = TextEditor()
editor.write('Halo, ini teks pertama')
editor.write('ini teks kedua')
editor.write('ini teks ketiga')

print(editor.undo())
print(editor.undo())
```

# **LATIHAN 1**

```python
#LATIHAN 1

class math:
  def __init__(self):
    self.history = []

  def operasi(self, operasi):
    self.history.append(operasi)

  def undo(self):
    if self.history:
      return f"Menghapus : {self.history.pop()}"
    return "Tidak ada yang dihapus"

  def show(self):
      return f"\nStack sekarang : {self.history}\n"

calc = math()
calc.operasi('1 + 7')
calc.operasi('10 * 5')
calc.operasi('5 / 1')
calc.operasi('4 - 2')

print(calc.show())
print(calc.undo())
print(calc.undo())
print(calc.show())
```

# **LATIHAN 2**

```python
#LATIHAN 2

from collections import deque
class history:
  def __init__(self):
    self.history = deque()

  def visit(self, fitur):
    self.history.append(fitur)

  def back(self):
    if self.history:
      print(f"Fitur saat ini : {self.history.pop()}")
    else:
      print('Tidak ada riwayat')

  def show(self):
    fitur = list(self.history)
    return f"Fitur yang diakses = {fitur}\n"

history = history()
history.visit('J17')
history.visit('Kamera')
history.visit('Efek')
history.visit('Edit')
history.visit('Tambah musik')

print(history.show())
history.back()
history.back()
```

# **LATIHAN 3**

```python
#LATIHAN 3

from queue import LifoQueue
class draft:
  def __init__(self):
    self.draft = LifoQueue()

  def tambah(self, draft):
    self.draft.put(draft)

  def show_first(self):
    if not self.draft.empty():
      return f"Draft terbaru : {self.draft.queue[-1]}"
    return "Tidak ada draft terbaru"

  def show(self):
    draft_list = list(self.draft.queue)
    return f"Draft Gmail saat ini : {draft_list}"

gmail = draft()
gmail.tambah('AB17')
gmail.tambah('AB17.rev')
gmail.tambah('AB17.rev(1)')
gmail.tambah('AB17.rev(2)')
print(gmail.show())
print(gmail.show_first())
```

# **LATIHAN 4**

```python
#LATIHAN 4

class film:
  def __init__(self):
    self.tiket = []

  def tambah(self, tiket):
    self.tiket.append(tiket)

  def kosong(self):
    return len(self.tiket) == 0

  def cek_tiket(self):
    if not self.kosong():
      return f'Mengecek tiket . . . : {self.tiket.pop()}'
    else:
      return "Tidak ada yang dicek"

  def show(self):
    return f"Tiket film : {self.tiket}\n"

nonton = film()
nonton.tambah('Tiket J17 #1')
nonton.tambah('Tiket J17 #2')
nonton.tambah('Tiket J17 #3')
nonton.tambah('Tiket J17 #4')
nonton.tambah('Tiket J17 #5')

print(nonton.show())
print(nonton.cek_tiket())
```
