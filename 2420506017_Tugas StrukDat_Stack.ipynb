

# **TUGAS**



```python
class hanoi:
  def __init__(self):
    self.tiang1 = []
    self.tiang2 = []
    self.tiang3 = []

  def tambah(self, cakram):
    self.tiang1.append(cakram)

  def pindah12(self):
    if self.tiang1 :
      push = self.tiang1[-1]
      self.tiang2.append(push)
      print(f'Memindahkan {push} ke Tiang 2')
      print(f'Tiang 2 = {self.tiang2}\n')
      self.tiang1.pop()
      return

  def pindah13(self):
    if self.tiang1 :
      push = self.tiang1[-1]
      self.tiang3.append(push)
      print(f'Memindahkan {push} ke Tiang 3')
      print(f'Tiang 3 = {self.tiang3}\n')
      self.tiang1.pop()
      return

  def pindah32(self):
    if self.tiang3 :
      push = self.tiang3[-1]
      self.tiang2.append(push)
      print(f'Memindahkan {push} ke Tiang 2')
      print(f'Tiang 2 = {self.tiang2}\n')
      self.tiang3.pop()
      return

  def pindah23(self):
    if self.tiang2 :
      push = self.tiang2[-1]
      self.tiang3.append(push)
      print(f'Memindahkan {push} ke Tiang 3')
      print(f'Tiang 3 = {self.tiang3}\n')
      self.tiang2.pop()
      return

  def show(self):
    return f"Tiang saat ini :\nTiang 1 = {self.tiang1}\nTiang 2 = {self.tiang2}\nTiang 3 = {self.tiang3}\n"


cakram = hanoi()
print('Selamat datang di game Tower of Hanoi!\nKesempatan langkah kamu : 10')
cakram.tambah('Cakram 1')
cakram.tambah('Cakram 2')
cakram.tambah('Cakram 3')
cakram.tambah('Cakram 4')
print(cakram.show())

cakram.pindah12()
print(cakram.show())
cakram.pindah13()
print(cakram.show())
cakram.pindah32()
print(cakram.show())
```
