```python
#UNDIRECTED GRAPH - ADJAENCY LIST

#make graph
graph = {
    'A' : ['B','C'],
    'B' : ['A','D','E'],
    'C' : ['A','F'],
    'D' : ['B'],
    'E' : ['B','F'],
    'F' : ['C','E']
}

#show graph
def show(graf):
  for i in graf:
    print(f'{i} -> {graf[i]}')

show(graph)
```

```python
#TRAVERSAL GRAPH - BFS

from collections import deque

def bfs(graph, start):
  visited = set()
  queue = deque([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end=' ')
      visited.add(vertex)
      queue.extend([i for i in graph[vertex] if i not in visited])

bfs(graph, 'A')
```

```python
#TRAVERSAL GRAPH - DFS

def dfs(graph, start, visited=None):
  if visited is None:
    visited = set()

  if start not in visited:
    print(start, end=' ')
    visited.add(start)
    for i in graph[start]:
      dfs(graph, i, visited)

dfs(graph, 'A')
```

```python
#LATIHAN PERSONAL

# graph = {
#     'A' : ['B','C'],
#     'B' : ['A','D','E'],
#     'C' : ['A','F'],
#     'D' : ['B'],
#     'E' : ['B','F'],
#     'F' : ['C','E']
# }

def count_neighbors(graph):
  result = {}
  for node in graph:
    tetangga = graph[node]
    jmlTetangga = len(tetangga)
    result[node] = jmlTetangga
  return result

result = count_neighbors(graph)
for i in result:
  print(f'Tetangga simpul {i} berjumlah {result[i]}')
```
