```python
#TUGAS 1
#MUHAMMAD ABDILLAH AZIZ
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

root = Node(20) #pengisian root
root.left = Node(8) #pengisian left child
root.right = Node(17) #pengisian right child
root.left.left = Node(1)  #pengisian left child dari node kiri
root.right.right = Node(1)  #pengisian right child dari node kanan

def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=" ")
    inorder(node.right)

print('Struktur Tree berdasarkan identitas saya :')
inorder(root)
```

```python
#TUGAS 2

class Node:    #membuat node baru
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

class BinarySearchTree: #memnyusun root dan child
  def __init__(self):
    self.root = None
  def insert(self, root, data):   #mengisi root dan child
    if root is None:  #mengecek apakah data root kosong
      return Node(data)
    if data < root.data:  #mengecek apakah nilai data yang diinput kurang dari nilai root
      root.left = self.insert(root.left, data)
    else: #kondisi jika nilai data yang diinput lebih dari nilai root
      root.right = self.insert(root.right, data)
    return root

def inorder(node):  #fungsi inorder transversal
  if node:  #mengecek apakah node kosong
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

def preorder(node): #fungsi preorder transversal
  if node:  #mengecek apakah node kosong
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

def postorder(node):  #fungsi postorder transversal
  if node:  #mengecek apakah node kosong
    postorder(node.left)
    postorder(node.right)
    print(node.data, end=' ')

def search(node, key):  #fungsi search node pada tree
  if node is None or node.data == key:  #mengecek apakah node kosong atau nilai data pada node sesuai nilai key
    return node
  if key < node.data: #mengecek apakah nilai key kurang dari nilai node
    return search(node.left, key)
  return search(node.right, key)

bat = BinarySearchTree()  #menjalankan class BinarySearchTree sebagai bat
root = None
data_list = [617, 1, 1, 65, 66, 14, 22] #mendeklarasikan data
for value in data_list:
  root = bat.insert(root, value) #memasukkan data pada tree
print('Inorder Traversal : ')
inorder(root) #menjalankan fungsi inorder transversal dengan data root
print('\n\nPreorder Traversal : ')
preorder(root)  #menjalankan fungsi preorder transversal dengan data root
print('\n\nPostorder Traversal : ')
postorder(root) #menjalankan fungsi postorder transversal dengan data root

#UJi
key = 25  #menguji pencarian nilai 25
result = search(root, key)  #menjalankan fungsi search dengan data dari root dan key 25
if result:  #mengecek apakah fungsi search berjalan dan mengambil hasil dari fungsi search
  print(f'\n\nApakah nilai {key} ada di Tree?\nYa, nilai {key} ditemukan dalam tree')
else:
  print(f'\n\nApakah nilai {key} ada di Tree?\nTidak, nilai {key} tidak ditemukan dalam tree')

key = 65  #menguji pencarian nilai 65
result = search(root, key)  #menjalankan fungsi search dengan data dari root dan key 65
if result:  #mengecek apakah fungsi search berjalan dan mengambil hasil dari fungsi search
  print(f'\nApakah nilai {key} ada di Tree?\nYa, nilai {key} ditemukan dalam tree')
else:
  print(f'\nApakah nilai {key} ada di Tree?\nTidak, nilai {key} tidak ditemukan dalam tree')
```
