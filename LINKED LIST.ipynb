# LATIHAN **2**

```python
#LATIHAN 2

class node:
  def __init__(self, folder):
    self.folder = folder
    self.next = None
    self.prev = None

class folder:
  def __init__(self):
    self.current = None

  def add(self, folder):
    new_folder = node(folder)
    if not self.current:
      self.current = new_folder
    else:
      new_folder.prev = self.current
      self.current.next = new_folder
      self.current = new_folder

  def show(self):
    if not self.current:
      print('\nTidak ada folder yang dikunjungi')
      return

    print('\nRiwayat folder : ')
    riwayat = self.current
    while riwayat.prev:
      riwayat = riwayat.prev

    n = 1
    while riwayat:
      print(f'{n}. {riwayat.folder}')
      riwayat = riwayat.next
      n+=1

  def kembali(self):
    if self.current and self.current.prev:
      self.current = self.current.prev
      print(f'\nAnda kembali ke : {self.current.folder}')
    else:
      print('\nTidak bisa kembali! Ini riwayat pertama')

  def maju(self):
    if self.current and self.current.next:
      self.current = self.current.next
      print(f'\nAnda maju ke : {self.current.folder}')
    else:
      print('\nTidak bisa maju! Ini riwayat terakhir')


history = folder()
jumlah_folder = int(input('Masukkan jumlah folder yang dikunjungi : '))
for i in range(1, jumlah_folder + 1):
  folder = input(f'Masukkan folder ke {i} : ')
  history.add(folder)

history.show()
while True:
  tombol = input('\nTekan B untuk kembali, Tekan F untuk maju, dan Tekan X untuk selesai : ').upper()
  if tombol == 'B':
    history.kembali()
  elif tombol == 'F':
    history.maju()
  elif tombol == 'X':
    break
  else:
    print('Tidak ada')
    break
```

```python
#LATIHAN 3

class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class LinkedList:
  def __init__(self):
    self.head = None

  def add_at_beginning(self, data):
    new_node = Node(data)
    if not self.head:
      self.head = new_node
      return
    temp = self.head
    while temp.next:
      temp = temp.next
    temp.next = new_node

  def add_at_end(self, data):
    new_node= Node(data)
    if not self.head:
      self.head = new_node
      return
    temp = self.head
    while temp.next:
      temp = temp.next
    temp.next = new_node

  def add_at_position(self, position, data):
    if position < 0:
      print('Posisi tidak valid!')
      return
    new_node = Node(data)
    if position == 0:
      self.add_at_beginning(data)
      return
    temp = self.head
    for _ in range(position - 1):
      if not temp:
        print('Posisi melebihi panjang dafta')
        return
      temp = temp.next
    new_node.next = temp.next
    temp.next = new_node

  def delete_at_beginning(self):
    if not self.head:
      print('Daftar kosong, tidak ada yang bisa dihapus')
      return
    self.head = self.head.next

  def delete_at_end(self):
    if not self.head:
      print('Daftar kosong, tidak ada yang bisa dihapus')
      return
    if not self.head.next:
      self.head = None
      return
    temp = self.head
    while temp.next.next:
      temp = temp.next
    temp.next = None

  def delete_at_position(self, position):
    if not self.head:
      print('Daftar kosong, tidak ada yang bisa dihapus')
      return
    if position == 0:
      self.head = self.head.next
      return
    temp = self.head
    for _ in range(position - 1):
      if not temp or not temp.next:
        print('Posisi tidak ditemukan!')
        return
      temp = temp.next
    temp.next = temp.next.next

  def search(self, key):
    temp = self.head
    while temp:
      if temp.data == key:
        print(f'Elemen {key} ditemukan dalam daftar!')
        return
      temp = temp.next
    print(f'Elemen {key} tidak ditemukan dalam daftar')

  def reverse(self):
    prev = None
    current = self.head
    while current:
      next_node = current.next
      current.next = prev
      prev = current
      current = next_node
    self.head = prev

  def display(self):
    if not self.head:
      print('Daftar kosong!')
      return
    temp = self.head
    while temp:
      print(temp.data, end=' -> ')
      temp = temp.next
    print('None')

ll = LinkedList()
num_elements = int(input('Masukkan jumlah awal elemen dalam daftar : '))
for i in range(num_elements):
  data = int(input(f'Masukkan data ke-{i} : '))
  ll.add_at_end(data)

#MENU OPERASI
while True:
  print('\nPilih operasi')
  print('1. Tambah elemen di awal')
  print('2. Tambah elemen di akhir')
  print('3. Tambah elemen di posisi tertentu')
  print('4. Hapus elemen di awal')
  print('5. Hapus elemen di akhir')
  print('6. Hapus elemen di posisi tertentu')
  print('7. Cari elemen dalam daftar')
  print('8. Balikkan daftar')
  print('9. Tampilkan daftar')
  print('0. Keluar')

  choice = int(input('\nMasukkan pilihan'))

  if choice == 1:
    data = int(input('Masukkan data yang ingin ditambahkan di awal : '))
    ll.add_at_beginning(data)
  elif choice == 2:
    data = int(input('Masukkan data yang ingin ditambahkan di akhir : '))
    ll.add_at_end(data)
  elif choice == 3:
    position = int(input('Masukkan posisi untuk menambahkan data : '))
    data = int(input('Masukkan data yang ingin ditambahkan : '))
    ll.add_at_position(position, data)
  elif choice == 4:
    ll.delete_at_beginning()
  elif choice == 5:
    ll.delete_at_end()
  elif choice == 6:
    position = int(input('Masukkan posisi elemen yang ingin dihapus : '))
    ll.delete_at_position(position)
  elif choice == 7:
    key = int(input('Masukkan elemen yang ingin dicari : '))
    ll.search(key)
  elif choice == 8:
    ll.reverse()
    print('Data dibalik!')
  elif choice == 9:
    ll.display()
  elif choice == 0:
    print('\nKeluar dari program')
    break
  else:
    print('\nPilihan tidak valid. COba lagi')
```

```python
#CIRCULLAR LIST

class SongNode:
  def __init__(self, title):
    self.title = title
    self.next = None

class Playlist:
  def __init__(self):
    self.current = None

  def add_song(self,title):
    new_song = SongNode(title)
    if not self.current:
      self.current = new_song
      self.current.next = self.current
      return
    else:
      temp = self.current
      while temp.next != self.current :
        temp = temp.next
      temp.next = new_song
      new_song.next = self.current

  def show_playlist(self):
    if not self.current:
      print('\nPlaylist kosong')
      return

    print('\nPlaylist anda : ')
    temp = self.current
    count = 1
    while True:
      print(f'{count}. {temp.title}')
      temp = temp.next
      count += 1
      if temp == self.current:
        break

  def next_song(self):
    if self.current:
      self.current = self.current.next
      print(f'\nMemutar : {self.current.title}')

#################################################
playlist = Playlist()
num_songs = int(input('\nMasukkan jumlah lagu dalam playlist : '))
for i in range(1, num_songs+1):
  title = input(f'Masukkan judul lagu ke - {i} : ')
  playlist.add_song(title)

playlist.show_playlist()

while True:
  command = input("\ntekan 'n' untuk lagu berikutnya, Tekan 'x' untuk keluar.\nMasukkan perintah : ").lower()
  if command == 'n':
    playlist.next_song()
  elif command == 'x':
    print('\nKeluar dari pemutar musik')
    break
  else:
    print('Perintah tidak ada')
```
