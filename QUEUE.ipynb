```python
#ANTRIAN PEMESANAN TIKET BIOSKOP

class Bioskop:
  def __init__(self):
    self.antrian = []

  def enqueue(self, tiket_id):
    self.antrian.append(tiket_id)

  def dequeue(self):
    if self.antrian:
      return self.antrian.pop(0)
    return None

  def front(self):
    return self.antrian[0] if self.antrian else None

  def rear(self):
    return self.antrian[-1] if self.antrian else None

  def is_empty(self):
    return len(self.antrian) == 0

  def size(self):
    return len(self.antrian)

  def akses(self):
    for i in range(len(self.antrian)):
      if '7' in self.antrian[i]:
        akses = self.antrian[i]
        self.antrian.pop(i)
        return f'Mengakses : {akses}'
    if self.antrian:
      akses = self.front()
      self.antrian.pop(0)
      return f'Mengakses : {akses}'
    return 'Antrian telah habis'

antrian = Bioskop()
while True:
  print(50*'=')
  print('|\tSelamat datang di Bioskop Fountaine!     |\n'+ 50*'=')
  print('| 1. Menginput kode pelanggan\t\t\t |')
  print('| 2. Menampilkan pelanggan pertama\t\t |')
  print('| 3. Menampilkan pelanggan terakhir\t\t |')
  print('| 4. Menampilkan jumlah antrian\t\t\t |')
  print('| 5. Akses antrian\t\t\t\t |')
  pilihan = int(input('Masukkan pilihan anda : '))
  print(50*"=")
  if pilihan == 1:
    jumlah = int(input('Masukkan jumlah pelanggan yang akan diinput : '))
    for i in range(jumlah):
      kode = input('Masukkan kode pelanggan : ')
      batas = antrian.size()
      if batas > 9 :
        print('Bioskop sudah penuh!\nTunggu sesi berikutnya')
        continue
      antrian.enqueue(kode)
      print(f'Antrian saat ini : {antrian.antrian}\n')
  elif pilihan == 2:
    print(f'\n<<<< Pelanggan pertama : {antrian.front()} >>>>\n')
  elif pilihan == 3:
    print(f'\n<<<< Pelanggan terakhir : {antrian.rear()} >>>>\n')
  elif pilihan == 4:
    print(f'\nAntrian saat ini : {antrian.antrian}\nJumlah antrian saat ini : {antrian.size()}\n')
  elif pilihan == 5 :
    print(antrian.akses())
  else:
    print('Perintah tidak ada')
    break
```

```python
#ANTRIAN PEMESANAN TIKET BIOSKOP

class Fotokopi:
  def __init__(self):
    self.id = []
    self.halaman = []

  def enqueue(self, order_id, num_pages):
    self.id.append(order_id)
    self.halaman.append(num_pages)

  def dequeue(self):
    if self.id and self.halaman:
      return f'Pelanggan saat ini : {self.id.pop(0)}\nJumlah halaman : {self.halaman.pop(0)}hlm\nMemproses....\nSELESAI!'
    return None

  def front(self):
    for i in range(len(self.halaman)):
      if self.halaman[i] < 10 or self.halaman[i] > 100:
        return f'<<< Pelanggan pertama : {self.id[i]} >>>\n<<< Jumlah halaman : {self.halaman[i]} >>>' if self.id and self.halaman else None
    return f'<<< Pelanggan pertama : {self.id[0]} >>>\n<<< Jumlah halaman : {self.halaman[0]} >>>' if self.id and self.halaman else None

  def rear(self):
    for i in range(len(self.halaman)-1,-1,-1):
      if not self.halaman[i] < 10 or not self.halaman[i] > 100:
       return f'<<< Pelanggan terakhir : {self.id[i]} >>>\n<<< Jumlah halaman : {self.halaman[i]} >>> ' if self.id and self.halaman else None
    return f'<<< Pelanggan terakhir : {self.id[-1]} >>>\n<<< Jumlah halaman : {self.halaman[-1]} >>> ' if self.id and self.halaman else None

  def is_empty(self):
    return len(self.id) == 0

  def size(self):
    return len(self.id)

  def akses(self):
    for i in range(len(self.halaman)):
      if self.halaman[i] < 10:
        fotokopi = self.id.pop(i)
        halaman = self.halaman.pop(i)
        return f'Pelanggan saat ini : {fotokopi}\nJumlah halaman : {halaman}hlm\nMemproses....\nSELESAI!\n'
      if self.halaman[i] > 100:
        print(f'Pelanggan saat ini : {self.id[i]}\nJumlah halaman : {self.halaman[i]}hlm')
        tahap1 = round(self.halaman[i]/2)
        tahap2 = self.halaman[i]-tahap1
        self.halaman[i] = tahap1
        print(f'Halaman terlalu banyak!\nMemproses tahap 1 : {self.halaman[i]} hlm\nTahap 1 Selesai!\n')
        self.halaman[i] = tahap2
        self.id.pop(i)
        return f'Memproses tahap 2 : {self.halaman.pop(i)} hlm\nTahap 2 Selesai!\n'
    if self.id and self.halaman:
      return self.dequeue()
    return 'Antrian telah habis'

antrian = Fotokopi()
while True:
  print(50*'=')
  print('|\tSelamat datang di Fotocopy Untidar!      |\n'+ 50*'=')
  print('| 1. Menginput kode pelanggan\t\t\t |')
  print('| 2. Menampilkan pelanggan pertama\t\t |')
  print('| 3. Menampilkan pelanggan terakhir\t\t |')
  print('| 4. Menampilkan jumlah antrian\t\t\t |')
  print('| 5. Akses antrian\t\t\t\t |')
  pilihan = int(input('Masukkan pilihan anda : '))
  print(50*"=")
  if pilihan == 1:
    jumlah = int(input('Masukkan jumlah pelanggan yang akan diinput : '))
    for i in range(jumlah):
      kode = input('Masukkan kode pelanggan : ')
      halaman = int(input('Masukkan jumlah halaman pelanggan : '))
      antrian.enqueue(kode, halaman)
      print(f'Antrian saat ini : {antrian.id}\n')
  elif pilihan == 2:
    print(antrian.front())
  elif pilihan == 3:
    print(antrian.rear())
  elif pilihan == 4:
    print(f'\nAntrian saat ini : {antrian.id}\nJumlah halaman ; {antrian.halaman}\nJumlah antrian saat ini : {antrian.size()}\n')
  elif pilihan == 5:
    print(antrian.akses())
  else:
    print('Perintah tidak ada')
    break
```

# LATIHAN **3**

```python
#ANTRIAN LAYANAN PARKIR MALL

class Parkir:
  def __init__(self):
    self.antrian = []

  def enqueue(self, vehicle_number):
    self.antrian.append(vehicle_number)

  def dequeue(self):
    if self.antrian:
      return self.antrian.pop(0)
    return f'Tidak ada kendaraan!'

  def front(self):
    for i in range(len(self.antrian)):
      if 'MA' in self.antrian[i]:
        return self.antrian[i]
    return self.antrian[0] if self.antrian else None

  def rear(self):
    for i in range(len(self.antrian)-1, -1,-1):
      if not 'MA' in self.antrian[i]:
        return self.antrian[i]
    return self.antrian[-1] if self.antrian else None

  def is_empty(self):
    return len(self.antrian) == 0

  def size(self):
    return len(self.antrian)

  def akses(self):
    for i in range(len(self.antrian)):
      if 'MA' in self.antrian[i]:
        keluar = self.antrian[i]
        self.antrian.pop(i)
        return f'Mengeluarkan kendaraan : {keluar}'
    if self.antrian:
      keluar = self.front()
      self.antrian.pop(0)
      return f'Meneluarkan kendaraan : {keluar}'
    return 'Parkiran kosong!'

antrian = Parkir()
while True:
  print(50*'=')
  print('|\tSelamat datang di Layanan Parkir Mall!   |\n'+ 50*'=')
  print('| 1. Menginput plat nomor kendaraan\t\t |')
  print('| 2. Menampilkan kendaraan pertama\t\t |')
  print('| 3. Menampilkan kendaraan terakhir\t\t |')
  print('| 4. Menampilkan jumlah kendaraan\t\t |')
  print('| 5. Akses antrian\t\t\t\t |')
  pilihan = int(input('Masukkan pilihan anda : '))
  print(50*"=")
  if pilihan == 1:
    jumlah = int(input('Masukkan jumlah kendaraan yang akan diinput : '))
    for i in range(jumlah):
      kode = input('Masukkan plat kendaraan : ')
      batas = antrian.size()
      if batas > 14 :
        print('Parkiran sudah penuh!\nTunggu kendaraan lain keluar!')
        continue
      antrian.enqueue(kode)
      print(f'Antrian saat ini : {antrian.antrian}\n')
  elif pilihan == 2:
    print(f'\n<<<< Kendaraan pertama : {antrian.front()} >>>>\n')
  elif pilihan == 3:
    print(f'\n<<<< Kendaraan terakhir : {antrian.rear()} >>>>\n')
  elif pilihan == 4:
    print(f'\nAntrian saat ini : {antrian.antrian}\nJumlah kendaraan saat ini : {antrian.size()}\n')
  elif pilihan == 5:
    print(antrian.akses())
  else:
    print('Perintah tidak ada')
    break
```

```python
#ANTRIAN PENGAMBILAN SERTIFIKAT KAMPUS

class Sertifikat:
  def __init__(self):
    self.id = []
    self.id_prestasi = []

  def enqueue(self,student_id):
    self.id.append(student_id)
    if len(self.id) > 20:
      temp_list = []
      for i in self.id:
        if "NAS" in i:
          self.id_prestasi.append(i)
        else:
          temp_list.append(i)
      self.id = temp_list

  def dequeue(self):
    if self.id_prestasi:
      return f'Mahasiswa saat ini : {self.id_prestasi.pop(0)}'
    if self.id:
      return f'Mahasiswa saat ini : {self.id.pop(0)}'
    return f'Tidak ada mahasiswa!'

  def front(self):
    for i in range(len(self.id)):
      if 'NAS'in self.id[i]:
        return f'<<< Mahasiswa pertama : {self.id[i]} >>>' if self.id else None
      else:
        return f'<<< Mahasiswa pertama : {self.id[0]} >>>' if self.id else None
    return f'<<< Mahasiswa pertama : {self.id_prestasi[0]} >>>' if self.id_prestasi else None

  def rear(self):
    for i in range(len(self.id)-1,-1,-1):
      if not 'NAS' in self.id[i]:
       return f'<<< Mahasiswa terakhir : {self.id[i]} >>>' if self.id else None
    return f'<<< Mahasiswa terakhir : {self.id[-1]} >>>' if self.id else None

  def is_empty(self):
    return len(self.id) == 0

  def size(self):
    return len(self.id) + len(self.id_prestasi)

  def akses(self):
    for i in range(len(self.id)):
      if 'NAS' in self.id[i]:
        return f'Mahasiswa saat ini : {self.id.pop(i)}\n'
    if self.id_prestasi:
        return f'Mahasiswa saat ini : {self.id_prestasi.pop(0)}\n'
    if self.id:
      return self.dequeue()
    return 'Antrian telah habis'

antrian = Sertifikat()
while True:
  print(57*'=')
  print('|   Selamat datang di Pengambilan Sertifikat Untidar!\t|\n'+ 57*'=')
  print('| 1. Menginput kode mahasiswa\t\t\t\t|')
  print('| 2. Menampilkan mahasiswa pertama\t\t\t|')
  print('| 3. Menampilkan mahasiswa terakhir\t\t\t|')
  print('| 4. Menampilkan jumlah antrian\t\t\t\t|')
  print('| 5. Akses antrian\t\t\t\t\t|')
  pilihan = int(input('Masukkan pilihan anda : '))
  print(57*"=")
  if pilihan == 1:
    jumlah = int(input('Masukkan jumlah mahasiswa yang akan diinput : '))
    for i in range(jumlah):
      kode = input('Masukkan kode mahasiswa : ')
      antrian.enqueue(kode)
      print(f'Antrian saat ini : {antrian.id}\n')
  elif pilihan == 2:
    print(antrian.front())
  elif pilihan == 3:
    print(antrian.rear())
  elif pilihan == 4:
    print(f'\nAntrian reguler : {antrian.id}\nAntrian prestasi : {antrian.id_prestasi}\nJumlah antrian saat ini : {antrian.size()}\n')
  elif pilihan == 5:
    print(antrian.akses())
  else:
    print('Perintah tidak ada')
    break
```
