```python
#TUGAS

class node:
  def __init__(self, nama):
    self.nama = nama
    self.next = None

class nama:
  def __init__(self):
    self.current = None

  def add(self, nama):
    new_nama = node(nama)
    if not self.current:
      self.current = new_nama
      self.current.next = self.current
      return
    else:
      temp = self.current
      while temp.next != self.current:
        temp = temp.next
      temp.next = new_nama
      new_nama.next = self.current

  def tugas(self):
    nama = self.current
    print('\n\nJadwal piket satpam selama dua minggu\n\n')
    for i in range(1, 15):
      print(f'Hari - {i} = {nama.nama}')
      nama = nama.next

# ----- Program Utama ----- #
satpam = nama()
jumlah = int(input('Masukkan jumlah satpam : '))
for i in range(1, jumlah+1):
  nama = input(f'Masukkan nama ke {i} : ')
  satpam.add(nama)

satpam.tugas()
```
